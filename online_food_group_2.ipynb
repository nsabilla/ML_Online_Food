## **1. INTRODUCTION**

### **Dataset**
Dataset yang digunakan pada project ini yaitu Online Food yang tersedia dari website kaggle melalui [link berikut]( https://www.kaggle.com/datasets/sudarshan24byte/online-food-dataset/data). Dataset ini terdiri dari 13 atribut dan 388 baris. Atribut yang menjadi variabel target adalah Output yang berisi "No" (tidak memesan makanan lagi) dan "Yes" (memesan makanan lagi).

### **Informasi Tentang Fitur-Fitur pada Dataset**
Demographic Information:

1. Age: Usia pelanggan.
2. Gender: Jenis kelamin pelanggan.
3. Marital Status:  Status perkawinan pelanggan.
4. Occupation:  Pekerjaan pelanggan.
5. Monthly Income: Pendapatan bulanan pelanggan.
6. Educational Qualifications: Kualifikasi pendidikan pelanggan.
7. Family Size: Jumlah individu dalam keluarga pelanggan.

Location Information:

1. Latitude: Lintang lokasi pelanggan.
2. Longitude: Bujur lokasi pelanggan..
3. Pin Code: Kode pin lokasi pelanggan.

Order Details:

1. Output: Status pesanan pelanggan saat ini (memesan makan lagi "Yes" dan tidak memesan makan lagi "No").
2. Feedback: Umpan balik yang diberikan oleh pelanggan setelah menerima pesanan.

## 4. Data Preprocessing
Bagian ini berisi proses penyiapan data berupa data cleaning sebelum dilakukan explorasi data lebih lanjut. Proses preprocessing dapat berupa mengisi missing values, menghapus kolom yang tidak dipakai, dan lain sebagainya.

### Menghapus kolom

```python
# Hapus attribute Unnamed: 12
data_cleaned = data.drop(columns=["Unnamed: 12"])
data_cleaned
```

### Filtering Data

```python
#Filtering data dengan target (Output) adalah 'No'
# data_cleaned[data_cleaned['Output']=='No']
```

```python
#Filtering data Output 'Yes' dan Gender adalah 'Female' serta Marital Status adalah 'Married'
data_cleaned[(data_cleaned['Output'] == 'Yes') &
              (data_cleaned['Gender'] == 'Female') &
              (data_cleaned['Marital Status'] == 'Married')]
```

- Analisa yang didapat dari hasil filtering data:
1. Mayoritas responden memiliki tingkat pendidikan lulusan sarjana atau lebih tinggi.

4. Rata-rata ukuran keluarga adalah sekitar 3-5 orang.
5. Responden tersebar di beberapa lokasi geografis yang berbeda, namun mayoritas berada di sekitar wilayah Bangalore (India) dengan beberapa kode pos yang berbeda.

```python
#Filtering data Output 'No' dan Gender adalah 'Male' serta Marital Status adalah 'Single'
data_cleaned[(data_cleaned['Output'] == 'No') &
              (data_cleaned['Gender'] == 'Male') &
              (data_cleaned['Marital Status'] == 'Single')]
```

### Menggunakan Pie Chart
Pie chart karena menampilkan distribusi berdasarkan data categorical

```python
# Menghitung jumlah pemesanan makanan berdasarkan jenis kelamin
gender_counts = data_cleaned["Gender"].value_counts()

# Membuat diagram pie
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(gender_counts, autopct='%1.1f%%', startangle=90)
plt.title('Siapa yang memesan makanan lebih banyak: Male or Female')
plt.axis('equal')  # Mengatur agar diagram menjadi lingkaran
# Menambahkan legenda
plt.legend(wedges, gender_counts.index, title="Gender", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

- Pada suatu jasa layanan food online presentasi tertinggi adalah laki-laki sebanyak 57.2%. Sedangkan, untuk perempuan sebanyak 42.8%
- Dengan persentase pengguna laki-laki yang lebih tinggi dibandingkan perempuan, layanan pemesanan makanan online sebaiknya mempertimbangkan strategi pemasaran yang lebih tersegmentasi dan fokus pada kebutuhan serta preferensi masing-masing gender. Ini akan membantu dalam meningkatkan kepuasan pelanggan dan memperluas basis pengguna layanan.

```python
# Menghitung jumlah pemesanan makanan berdasarkan status pernikahan
martial_status_counts = data_cleaned["Marital Status"].value_counts()

plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(martial_status_counts, autopct='%1.1f%%', startangle=45)
plt.title('Siapa yang memesan makanan lebih banyak: Married or Singles')
# Mengatur agar diagram menjadi lingkaran
plt.axis('equal')
# Menambahkan legenda
plt.legend(wedges, martial_status_counts.index, title="Marital Status", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

- Presentase yang tinggi dari pelanggan yang statusnya "Single" bisa menunjukkan bahwa layanan pemesanan makanan online lebih diminati oleh individu yang tinggal sendiri.

```python
# Menghitung jumlah pemesanan makanan berdasarkan status pernikahan dan gender
gender_marital_counts = data_cleaned.groupby(['Gender', 'Marital Status']).size().unstack()

# Membuat diagram pie untuk setiap gender
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

genders = gender_marital_counts.index

for i, gender in enumerate(genders):
    ax = axes[i]
    marital_status_counts = gender_marital_counts.loc[gender]
    wedges, texts, autotexts = ax.pie(marital_status_counts, autopct='%1.1f%%', startangle=90)

    ax.set_title(f'Siapa yang memesan makanan lebih banyak ({gender})')

    ax.legend(wedges, marital_status_counts.index, title="Marital Status", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Mengatur layout agar tidak terpotong
plt.tight_layout()
plt.show()

```

- Data menunjukkan bahwa lebih banyak pemesanan makanan online dilakukan oleh laki-laki yang statusnya "Single" dibandingkan dengan perempuan yang statusnya "Single". Hal ini mungkin disebabkan oleh preferensi atau kebiasaan konsumsi yang berbeda antara laki-laki dan perempuan yang tinggal sendiri.
- Sebaliknya, untuk pelanggan yang sudah menikah, terlihat bahwa lebih banyak pemesanan makanan online dilakukan oleh perempuan dibandingkan dengan laki-laki. Hal ini bisa disebabkan oleh peran tradisional dalam keluarga di mana perempuan lebih sering bertanggung jawab atas mempersiapkan makanan.

```python
# Menghitung jumlah pemesanan makanan berdasarkan tingkat pendapatan bulanan
monthly_income_counts = data_cleaned["Monthly Income"].value_counts()

# Membuat diagram pie
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(monthly_income_counts, autopct='%1.1f%%', startangle=90)
plt.title('Menganalisis Perilaku Pemesanan Makanan di Seluruh Tingkat Pendapatan')
plt.axis('equal')  # Mengatur agar diagram menjadi lingkaran
plt.legend(wedges, monthly_income_counts.index, title="Monthly Income", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

- Presentase pelanggan untuk online food ini orang-orang yang tidak memiliki pendapatan per bulan. Kemungkinan, layanan ini menawarkan kemudahan akses dan berbagai promosi yang membuatnya lebih terjangkau bagi mereka yang tidak memiliki pendapatan per bulan.

```python
# Menghitung jumlah pemesanan makanan berdasarkan kualifikasi pendidikan dan Monthly Income
qualification_income_counts = data_cleaned.groupby(["Educational Qualifications", "Monthly Income"]).size().unstack()

# Membuat bar plot
qualification_income_counts.plot(kind='bar', figsize=(12, 5))
plt.xlabel('Educational Qualifications')
plt.ylabel('Count')
plt.title('Pola pemesanan makanan berdasarkan kualifikasi pendidikan dan pendapatan per bulan')
plt.xticks(rotation=45)  # Mengatur rotasi label pada sumbu x
plt.legend(title='Monthly Income')
plt.show()
```

- Pemesanan makanan berdasarkan kualifikasi pendidikan dan pendapatan per bulan menanyakan bahwa kualifikasi pendiidkan Graduate lebih setara pendapatan per bulan. Pola ini dapat membantu kita mengidentifikasi kelompok pelanggan mana yang paling aktif dalam memesan makanan

### Menggunakan Bar Chart
Sama seperti pie chart untuk menampilkan distribusi berdasarkan data categorical

```python
# Filtering untuk visualisasi EDA
buying_food_again = data_cleaned.query("Output == 'Yes'")
buying_food_again.head()
```

```python
# Data umpan balik dari keseluruhan data
feedback_counts = data_cleaned["Feedback"].value_counts()

# Data umpan balik dari data yang hanya melakukan pembelian makanan ulang
feedback_counts_again = buying_food_again["Feedback"].value_counts()

# Membuat dua bar plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Bar plot untuk keseluruhan data
ax1.bar(feedback_counts.index, feedback_counts.values, color=['skyblue', 'salmon'])
ax1.set_title('Positive vs Negative Feedback')
ax1.set_xlabel('Feedback')
ax1.set_ylabel('Count')
ax1.set_xticks(feedback_counts.index)
ax1.set_xticklabels(['Positive', 'Negative'])
ax1.set_ylim(0, max(feedback_counts.values) + 10)

# Bar plot untuk data yang hanya melakukan pembelian makanan ulang
ax2.bar(feedback_counts_again.index, feedback_counts_again.values, color=['skyblue', 'salmon'])
ax2.set_title('Positive vs Negative Feedback berdasarkan pembelian makanan berulang')
ax2.set_xlabel('Feedback')
ax2.set_ylabel('Count')
ax2.set_xticks(feedback_counts_again.index)
ax2.set_xticklabels(['Positive', 'Negative'])
ax2.set_ylim(0, max(feedback_counts_again.values) + 10)

plt.tight_layout()
plt.show()
```

- Bar chart (kiri) menghitung Feedback "positive" dan "negative". Terlihat bar Feedback "positive" ini banyak diberikan oleh pelanggan.
- Dan, bar chart (kanan) Feedback "positive" dan "negative" terhadap variabel target Output "Yes" dan Feedback "positive" banyak juga diberikan oleh pelanggan
- Dari kedua bar chart diatas ini mengindikasikan bahwa mayoritas pelanggan memberikan feedback yang positif atas layanan atau produk yang mereka terima.
